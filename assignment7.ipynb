```python
class Vehicle:
    def __init__(self):
       self.x = 0
       self.y = 0
    def moveUp(self):
       self.y+=1 
    def moveDown(self):
       self.y-=1
    def moveRight(self):
       self.x+=1
    def moveLeft(self):
        self.x-=1
    def __str__(self):
        return '('+str(self.x)+' , '+str(self.y)+')'

class Vehicle2010(Vehicle):
    def moveLowerLeft(self):
      self.x-=1
      self.y-=1
     
    def moverlowerRight(self):
       self.x+=1
       self.y-=1

    def moveUpperRight(self):
      self.x+=1
      self.y+=1
    def moveUpperLeft(self):
      self.x-=1
      self.y+=1     
    def equals(self ,x):
      if self.x == x.x and self.y==x.y:
        return True
      else:
        return False  
       


#write your code here
print('Part 1')
print('------')
car = Vehicle()
print(car)
car.moveUp()
print(car)
car.moveLeft()
print(car)
car.moveDown()
print(car)
car.moveRight()
print(car)
print('------')
print('Part 2')
print('------')
car1 = Vehicle2010()
print(car1)
car1.moveLowerLeft()
print(car1)
car2 = Vehicle2010()
car2.moveLeft()
print(car1.equals(car2))
car2.moveDown()
print(car1.equals(car2))
```

```python
class Tournament:
  def __init__(self,name='Default'):
    self.__name = name
  def set_name(self,name):
    self.__name = name
  def get_name(self):
     return self.__name
  

class Cricket_Tournament(Tournament):
  def __init__(self, *num):
    if len(num)==0:
      super().__init__()
      self.t_ype="No type"
      self.no_teams=0
    elif len(num)==3:
      super().__init__(num[0])
      self.t_ype=num[2]
      self.no_teams=num[1]

  def detail(self):
      x="Tournament Name:" +self.get_name() + "\n" +"Number of Teams: "+ str(self.no_teams) + "\n" + "Type:" + self.t_ype
      return x

class Tennis_Tournament(Tournament):
   def __init__(self ,*num):
     super().__init__(num[0])
     self.no_players=num[1]

   def detail(self):
      x="Tournament Name:"+ self.get_name() +"\n"+ "Number of Players:" + str(self.no_players)
      return x



#write your code here
ct1 = Cricket_Tournament()
print(ct1.detail())
print("-----------------------")
ct2 = Cricket_Tournament("IPL",10,"t20")
print(ct2.detail())
print("-----------------------")
tt = Tennis_Tournament("Roland Garros",128)
print(tt.detail())
```

```python
class Shape:
    def __init__(self, name='Default', height=0, base=0):
            self.area = 0
            self.name = name
            self.height = height
            self.base = base
    def get_height_base(self):
       return "Height: "+str(self.height)+",Base: "+str(self.base)



class triangle(Shape):
  def __init__(self, name="Default" , height=0, base=0 ):
   super().__init__(name , height , base)
   
  def calcArea(self):
    if self.name== "Default":
      self.area=0

    elif self.name=='Triangle' :
      self.area= (self.base*self.height)/2

    


  def printDetail(self):
    
      x= "Name:" + self.name + "\n" + str(self.get_height_base()) + "\n" + "Area " + str(self.area) 
      return x

    
class trapezoid(Shape):
     def __init__(self, name="Default" , height=0, base=0 , base2=0):
      super().__init__(name , height , base)
      self.base2=base2 
   
   
     def calcArea(self):      
   
         self.area= ((self.base+self.base2)/2)*self.height
     
     def printDetail(self):
    
        x= "Name:" + self.name + "\n" + str(self.get_height_base())+" "+ "Side_A:"+ str(self.base2) + "\n" + "Area " + str(self.area) 
        return x

     
     
     

#write your code here









tri_default = triangle()
tri_default.calcArea()
print(tri_default.printDetail())
print('--------------------------')
tri = triangle('Triangle', 10, 5)
tri.calcArea()
print(tri.printDetail())
print('---------------------------')
trap = trapezoid('Trapezoid', 10, 6, 4)
trap.calcArea()
print(trap.printDetail())
```
